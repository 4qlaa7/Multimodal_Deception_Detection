**Reading and Saving Data**

```python
import csv
import numpy as np
from read import Read_Data,Padding
Facial_Features1 = '/content/drive/MyDrive/Graduation Project/Features/Facial_Deceptive_features_dlib.csv'
Audio_Features1 = '/content/drive/MyDrive/Graduation Project/Features/Deceptive_Audio_Features.csv'
Trans_Features1 = '/content/drive/MyDrive/Graduation Project/Features/Deceptive_Transcript_Features.csv'



Facial_Features2 = '/content/drive/MyDrive/Graduation Project/Features/Facial_Truthful_features_dlib.csv'
Audio_Features2 = '/content/drive/MyDrive/Graduation Project/Features/Truthful_Audio_Features.csv'
Trans_Features2 = '/content/drive/MyDrive/Graduation Project/Features/Truthful_Transcript_Features.csv'


print("Reading Train Data.......")
TrainF = []
TrainL = []
for i in range(40):
  facial1 = Read_Data(i,Audio_Features1)
  text1 = Read_Data(i,Trans_Features1)
  arr1 = np.concatenate((facial1,text1))
  np.array(arr1)
  TrainF.append(arr1)
  TrainL.append(1)


  facial2 = Read_Data(i,Audio_Features2)
  text2 = Read_Data(i,Trans_Features2)

  arr2 = np.concatenate((facial2,text2))
  np.array(arr2)
  TrainF.append(arr2)
  TrainL.append(0)


print(len(TrainF))
print(len(TrainL))

print("Reading Validation Data.......")


ValF = []
ValL = []
for i in range(40,50):
  facial1 = Read_Data(i,Audio_Features1)
  text1 = Read_Data(i,Trans_Features1)

  arr1 = np.concatenate((facial1,text1))
  np.array(arr1)
  ValF.append(arr1)
  ValL.append(1)
  facial2 = Read_Data(i,Audio_Features2)
  text2 = Read_Data(i,Trans_Features2)

  arr2 = np.concatenate((facial2,text2))
  np.array(arr2)
  ValF.append(arr2)
  ValL.append(0)

print(len(ValF))
print(len(ValL))
```

```python
TrainF = Padding(TrainF)
ValF = Padding(ValF)

TrainF = np.asarray(TrainF).astype(np.float32)
TrainL = np.asarray(TrainL).astype(np.float32)
ValF = np.asarray(ValF).astype(np.float32)
ValL = np.asarray(ValL).astype(np.float32)


print("TrainF shape:", TrainF.shape)
print("TrainL shape:", TrainL.shape)
print("TestF shape:", ValF.shape)
print("TestL shape:", ValL.shape)
print("TrainF type:", TrainF.dtype)
print("TrainL type:", TrainL.dtype)
print("TestF type:", ValF.dtype)
print("TestL type:", ValL.dtype)
```

```python
np.savez('A+T.npz', TrainF=TrainF, TrainL=TrainL,ValF = ValF , ValL = ValL)
```
